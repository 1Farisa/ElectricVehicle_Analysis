# Electric Vehicles Market Size Analysis


### Market Size Analysis is the process of estimating the potential sales for a product or service within a particular market segment. In the context of electric vehicles (EVs), it involves assessing the total volume of EV registrations to understand the growth of the market, forecast future trends, and help stakeholders make informed decisions regarding production, infrastructure development, and policy-making.


Dataset Link : https://statso.io/wp-content/uploads/2024/03/EV-data.zip

The provided dataset contains the following columns, each representing different aspects of the electric vehicle (EV) population in the United States:

- VIN (1-10): Partial Vehicle Identification Number.
- County: The county in which the vehicle is registered.
- City: The city in which the vehicle is registered.
- State: The state in which the vehicle is registered. It appears that this dataset may be focused on Washington (WA) state.
- Postal Code: The postal code where the vehicle is registered.
- Model Year: The year of the vehicle model.
- Make: The manufacturer of the vehicle.
- Model: The model of the vehicle.
- Electric Vehicle Type: The type of electric vehicle, e.g., Battery Electric Vehicle (BEV).
- Clean Alternative Fuel Vehicle (CAFV) Eligibility: Eligibility status for clean alternative fuel vehicle programs.
- Electric Range: The maximum range of the vehicle on a single charge (in miles).
- Base MSRP: The Manufacturer’s Suggested Retail Price.
- Legislative District: The legislative district where the vehicle is registered.
- DOL Vehicle ID: Department of Licensing Vehicle Identification.
- Vehicle Location: Geographic coordinates of the vehicle location.
- Electric Utility: The electric utility service provider for the vehicle’s location.
- 2020 Census Tract: The census tract for the vehicle’s location.

```python
electric_range = df['Electric Range']
```

```python
# Analyze the electric range distribution using a histogram with a KDE curve.
plt.figure(figsize=(10, 6))
sns.histplot(electric_range, bins=30, kde=True, color='skyblue')
plt.title("Electric Range Distribution Among Vehicles")
plt.xlabel("Electric Range (miles)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Visualizing the distribution of EV makes using a horizontal bar chart.
plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='Make', order=df['Make'].value_counts().index, palette='viridis')
plt.title("Distribution of Electric Vehicle Makes")
plt.xlabel("Count of Vehicles")
plt.ylabel("Make")
plt.show()
```

- Highlights manufacturers leading the market.
- Lesser counts for some brands could imply niche markets or limited adoption.
- Dominance by certain makes suggests brand reliability or popularity.

```python
# Compute the average electric range by model year and visualize it with a line plot.
average_range_by_year = df.groupby('Model Year')['Electric Range'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Model Year', y='Electric Range', data=average_range_by_year, marker='o', color='green')
plt.title('Average Electric Range by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average Electric Range (miles)')
plt.grid(True)
plt.show()
```

- A steady increase reflects technological advancements in battery capacity and efficiency.
- Sudden spikes might indicate breakthrough innovations.
- Increasing ranges are key to consumer adoption, addressing range anxiety.

```python
ev_makedistribution = df['Make'].value_counts().head(10)
ev_makedistribution
```

```python
# Identify the top 10 EV makes by number of registrations and visualize them with a bar chart.
plt.figure(figsize=(12, 6))
sns.barplot(x=ev_makedistribution.values, y=ev_makedistribution.index, palette="cubehelix")
plt.title('Top 10 Popular EV Makes')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Make')
plt.tight_layout()
plt.show()
```

- The top three manufacturers dominate the market, leaving little room for competitors.
- Popularity could correlate with brand reputation or extensive model portfolios.
- Leaders like Tesla may indicate consumer trust in performance and technology.


```python
# Focus on the top 3 EV makes by registration count.
top_3_makes = ev_makedistribution.head(3).index



# Filter the dataset for the top 3 makes.
top_makes_data = df[df['Make'].isin(top_3_makes)]
top_makes_data

# Group by make and model to find the distribution of vehicle registrations among top models.
ev_model_distribution_top_makes = top_makes_data.groupby(['Make', 'Model']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')

ev_model_distribution_top_makes
```

```python
top_models = ev_model_distribution_top_makes.head(10)

# Visualize the top models within the top 3 makes using a bar chart.
plt.figure(figsize=(12, 8))
sns.barplot(x='Number of Vehicles', y='Model', hue='Make', data=top_models, palette="viridis")
plt.title('Top Models in Top 3 Makes by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.tight_layout()
plt.show()
```

- Certain models significantly outperform others, showcasing consumer trust.
- Popular models might combine affordability, range, and performance.
- These models likely represent a strong value proposition for buyers.

```python
# Compute the average electric range for each model under the top makes.
average_range_by_model = top_makes_data.groupby(['Make', 'Model'])['Electric Range'].mean().sort_values(ascending=False).reset_index()

top_range_models = average_range_by_model.head(10)
# Visualize the top 10 models with the highest average electric range.
plt.figure(figsize=(12, 8))
barplot = sns.barplot(x='Electric Range', y='Model', hue='Make', data=top_range_models, palette="cool")
plt.title('Top 10 Models by Average Electric Range in Top Makes')
plt.xlabel('Average Electric Range (miles)')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.show()
```

- High-range models are likely luxury or premium offerings.
- Models with top ranges reflect the forefront of EV innovation.
- Consumers seeking long-range EVs gravitate toward these models, indicating a strong relationship between range and desirability.
